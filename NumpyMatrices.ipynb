#### Dot Product (Matris Çarpımı) Kuralı :

1.matrisin sütun (column) sayısı ile 2. matrisin satır (row) sayısı EŞİT olmalıdır.

Sonuç matrisinin boyutu: (1. matrisin satır sayısı, 2. matrisin sütun sayısı) olur.

```python
first_matrix = np.array([[10,20,30]])
```

```python
first_matrix.shape
```

```python
second_matrix = np.array([[2,3],[2,3],[2,3]])
```

```python
second_matrix.shape
```

```python
# first_matrix * second_matrix  : Hata alırız Çarpma işlemi gerçekleşmez
```

```python
result_matrix = first_matrix.dot(second_matrix)
```

```python
result_matrix
```

```python
result_matrix.shape
```
